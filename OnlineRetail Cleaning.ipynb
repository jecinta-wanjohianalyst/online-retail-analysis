```python
df.columns = df.columns.str.strip()
df["Description"] = df["Description"].str.strip()
df["Country"] = df["Country"].str.strip()
```

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
```

```python
df.dtypes
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
# Missing Description
df["Description"] = df["Description"].fillna("Unknown Product")
```

```python
# Missing CustomerID
df["CustomerID"] = df["CustomerID"].fillna(0)
```

```python
# Convert CustomerID back to integer
df["CustomerID"] = df["CustomerID"].astype(int)
```

```python
# Checking negative quantities and removing them
df[df["Quantity"] < 0]
(df["Quantity"] < 0).sum()
```

```python
df = df[df["Quantity"] > 0]
```

```python
# Checking negative prices and removing them
df[df["UnitPrice"] < 0]
df = df[df["UnitPrice"] > 0]
```

```python
# Canceled Orders
df[df["InvoiceNo"].astype(str).str.startswith("C")]
df["InvoiceNo"].astype(str).str.startswith("C").sum()
df = df[~df["InvoiceNo"].astype(str).str.startswith("C")]
```

```python
# Creating revenue column
df["Revenue"] = df["Quantity"] * df["UnitPrice"]
```

```python
df.to_csv("OnlineRetail_Cleaned.csv", index=False)
```
